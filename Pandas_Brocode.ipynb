#Series
pandas 1-dimensional labeled array that can hold any data type [think series like a single column in a spreadsheet]


```python
data=[100,200,300]
series=pd.Series(data,index=["a","b","c"])
print(series)

#loc : location by label

print(series.loc['a'])
print(series.loc['c'])

#series.loc['d'] Key error as d dont exist

#iloc : integer location
print(series.iloc[2])
```

#Filter by value

```python
data=[100,102,104,200,202]
series=pd.Series(data,index=["a","b","c","d","e"])
print(series[series>=200])
```

#DataFrames
A tabular data structure with rows and columns.(2dimensional)  [similar to excel spredsheet]


```python
import pandas as pd
data={
    'Name':['Ajita','Jake','Suno'],
    'Age':[25,23,21],
      }
df=pd.DataFrame(data,index=['Employee1','Employee2','Employee3'])
print(df)

#Acessing single row through loc
#df.loc['Employee1']

#Acessing single row through iloc
#df.iloc[2]

#Add new column in dataframe
df['Department']=['ML','Data Analyst','Cook']
#print(df)

#Add a new row
#new_row=pd.DataFrame([{'Name':'Sandy','Age':27,'Department':'Engineer'}],index=['Employee4'])
#df=pd.concat([df,new_row])


#Multiple rows add
new_row=pd.DataFrame([{'Name':'Sandy','Age':27,'Department':'Engineer'},
                      {'Name':'Jack','Age':25,'Department':'ML'},
                      {'Name':'Radha','Age':22,'Department':'AI'}],
                     index=['Employee4','Employee5','Employee6'])
df=pd.concat([df,new_row])
print(df)


```

#Aggregation

reduces a set of values to a single summary value

used to summarize and analyze data  

often used with the groupby() function



```python
import pandas as pd
df=pd.read_csv('Salary Data.csv')

#whole dataframe
#print(df.mean(numeric_only=True))
#print(df.sum(numeric_only=True))
#print(df.min(numeric_only=True))
#print(df.max(numeric_only=True))
#print(df.count())

#Single column
#print(df['Age'].mean())
#print(df['Age'].min())
#print(df['Age'].max())
#print(df['Age'].sum())
#print(df['Salary'].count())

group=df.groupby('Education Level')
print(group)  #return object so we use groupby with agregate function
print(group['Age'].mean())
print(group['Salary'].min())
```
