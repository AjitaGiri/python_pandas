# Seaborn

Univariate analysis: single feature
Bivariate analysis: two features

if we have mutliple features on dataset, we can visualize through seaborn

Distribution plots
. distplot: bivariate analysis
. joinplot: 
. pairplot: 

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

in above dataset, tip is dependent feature and rest of them are independent features.

### Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. Imp in feature selection

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pairplot
A Pair plot also known as scatterplot, in which one variable in the same data row is matched with another variable's value.

when we have multiple independent variables in dataset, we can use pairplot to visualize the data

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

### Dist plot
helps us to check the distribution of the column features

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=False,bins=10)
```

### Categorical Plots
seaborn also helps us in doing the analysis on Categorical data points.

. boxplot
. violinplot
. countplot
. bar plot


```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,color='pink')
```

```python
#Viloin plot: combination of box plot and kernel density estimation
sns.violinplot(x='total_bill',y='day',data=df)
```
