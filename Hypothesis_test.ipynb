# Hypothesis Testing

### A manufacturer claims that the average weight of a bag of potato chips is 150g. A sample of 25 bages is taken and the average weight is found to be 148g,with a standard deviation of 5g. Test the manufacturer's claim using a one-tailed t-test with a significance level of 0.05

Solution:

step1
Ho: µ =150
Ha: µ < 150

step2
level of significance α =0.05

step 3
tcalc= (sample_mean-µ)/ (s/√n)

step4
ttab= tα/2,n-1

```python
#step3
import numpy as np
import scipy.stats as st
sample_mean=148
µ=150
s=5
n=25

t_cal= (sample_mean-µ)/(s/np.sqrt(n))
print(t_cal)
```

```python
t_tab=st.t.ppf(0.05,24)

if t_cal < t_tab:
    print("reject Ho")
else:
    print("do not reject Ho")
```

step 5

Since t_cal< t_tab so we reject Ho.
Hence the tje manufactor claim is incorrect.

##  Q2
Suppose the IQ in a certain population is normally distributed with a mean of µ=100 and standard deviation of σ=15.

A researcher wants to know if a new drug affects IQ levels, so he recruits 18 patients to try it and record their IQ levels. Determine if the new drug cause a significant difference in IQ levels.

data=[88,92,94,96,97,97,99,99,105,109,109,109,110,112,112,113,114,115]

even though sample size is <30,Since population standard deviation is known so we can use z_test

step1
Ho: µ=100
H1: µ!=100

step 2
α=0.05

step 3
z_cal= (sample_mean-µ)/(σ/√n)

step4
z_tab


```python
import numpy as np
import scipy.stats as sts

#step3
data=np.array([88,92,94,96,97,97,99,99,105,109,109,109,110,112,112,113,114,115])

sample_mean=np.mean(data)
n=18
σ=15
µ=100

z_cal=(sample_mean-µ)/(σ/np.sqrt(n))
print(z_cal)
```

```python
z_tab=sts.norm.ppf(1-0.025)
#print(z_tab)

if z_cal > z_tab:
    print("reject Ho")
else:
    print("do not reject Ho")
```

```python
#we can also use p_value to make decisions

from statsmodels.stats.weightstats import ztest

α=0.05
data=[88,92,94,96,97,97,99,99,105,109,109,109,110,112,112,113,114,115]

z_stat, p_value = ztest(data, value=100)

print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < α:
    print("reject H0")
else:
    print("do not reject H0")
```

### compare p_value with α.
if p_value< α,reject H0
if p_value >= α, dont reject H0
