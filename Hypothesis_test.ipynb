# Hypothesis Testing

### A manufacturer claims that the average weight of a bag of potato chips is 150g. A sample of 25 bages is taken and the average weight is found to be 148g,with a standard deviation of 5g. Test the manufacturer's claim using a one-tailed t-test with a significance level of 0.05

Solution:

step1
Ho: µ =150
Ha: µ < 150

step2
level of significance α =0.05

step 3
tcalc= (sample_mean-µ)/ (s/√n)

step4
ttab= tα/2,n-1

```python
#step3
import numpy as np
import scipy.stats as st
sample_mean=148
µ=150
s=5
n=25

t_cal= (sample_mean-µ)/(s/np.sqrt(n))
print(t_cal)
```

```python
t_tab=st.t.ppf(0.05,24)

if t_cal < t_tab:
    print("reject Ho")
else:
    print("do not reject Ho")
```

step 5

Since t_cal< t_tab so we reject Ho.
Hence the tje manufactor claim is incorrect.

##  Q2
Suppose the IQ in a certain population is normally distributed with a mean of µ=100 and standard deviation of σ=15.

A researcher wants to know if a new drug affects IQ levels, so he recruits 18 patients to try it and record their IQ levels. Determine if the new drug cause a significant difference in IQ levels.

data=[88,92,94,96,97,97,99,99,105,109,109,109,110,112,112,113,114,115]

even though sample size is <30,Since population standard deviation is known so we can use z_test

step1
Ho: µ=100
H1: µ!=100

step 2
α=0.05

step 3
z_cal= (sample_mean-µ)/(σ/√n)

step4
z_tab


```python
import numpy as np
import scipy.stats as sts

#step3
data=np.array([88,92,94,96,97,97,99,99,105,109,109,109,110,112,112,113,114,115])

sample_mean=np.mean(data)
n=18
σ=15
µ=100

z_cal=(sample_mean-µ)/(σ/np.sqrt(n))
print(z_cal)
```

```python
z_tab=sts.norm.ppf(1-0.025)
#print(z_tab)

if z_cal > z_tab:
    print("reject Ho")
else:
    print("do not reject Ho")
```

```python
#we can also use p_value to make decisions

from statsmodels.stats.weightstats import ztest

α=0.05
data=[88,92,94,96,97,97,99,99,105,109,109,109,110,112,112,113,114,115]

z_stat, p_value = ztest(data, value=100)

print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < α:
    print("reject H0")
else:
    print("do not reject H0")
```

### compare p_value with α.
if p_value< α,reject H0
if p_value >= α, dont reject H0


# Chi-square test

A study was conducted to investigate whether there is a relationship brtween gender and the preferred genre of music. A sample of 235 people was selected. Test at a 5% level of sifnificance wther there is a significant association between gender and music preference.



col=pop,hiphop,classical,rock

male= 40,45,25,10

female=35,30,20,30

H0: There is no association between gender and the preferred genre of music

H1: there is association between gender and the preferred genre of music

α=0.05

```python
import numpy as np
import scipy.stats as sts
import pandas as pd

male=np.array([40,45,25,10])
female=np.array([35,30,20,30])

male_sum=np.sum(male)
female_sum=np.sum(female)

sum_row=np.array([male_sum,female_sum])
print(sum_row)
sum= male+female
  
```

```python
expected_freq=[]
for i in sum_row:
        for j in sum:
                value=(i*j)/235
                expected_freq.append(value)
print(expected_freq)
```

```python
observed_data=np.array([40,45,25,10,35,30,20,30])

result=np.sum((np.square(observed_data-expected_freq))/expected_freq)
print(result)
```

```python
# α,(row-1)*(col-1)
# 0.05,(2-1)*(4-3)
chisquare_tab=- 7.815
chisquare_cal=result
```

Decision
chisquare_cal > chisquare_tab, we reject H0.

Hence we conclude that there is association between gender and the preferred genre of music
