```python
# Write a function that takes a list of numbers and returns the sum.

def sum_num(num):
    total=0
    for i in num:
        total+=i
    return total

sum_num([1,2,3,4,5])
```

```python
# reverse a string
def rev_string(str_var):
    result= str_var[::-1]
    return result

rev_string('apple')
```

```python
def check_palindrom(num):
    rev_num= num[::-1]
    if num==rev_num:
        return True
    else:
        return False
    
check_palindrom('apple')
check_palindrom('515')
```

```python
# remove duplicates from a list

def remove_duplicate(lst):
    unique_list=[]
    for i in lst:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

remove_duplicate([1,2,2,2,3])
```

```python
def remove_duplicate(lst):
    set1=set()
    for i in lst:
        set1.add(i)
    return list(set1)

remove_duplicate([1,2,2,2,3,4])
```

```python
## fibonacci series for n-th number using recursion

def fibonacii(n):
    if n<=1:
        return n
    else:
        return fibonacii(n-1)+fibonacii(n-2)
    
fibonacii(5)
    
```

```python
# decorator

def my_decorator(func):
    def wrapper(*args,**kwargs):
        result=func(*args,**kwargs)
        return result
    return wrapper

@my_decorator
def greet(gretting):
    return f"{gretting},Narottam"

greet("Namaste")
```

```python
#Given a list of numbers, write a Python function to find the second highest number.
def second_highst(num_ls):
    inpt=list(set(num_ls)) #remove duplicate
    inpt.sort() #sort in asc order
    return inpt[-2] #return 2nd last

second_highst([23,23,56,78,2])
```

```python
# factorial of number using recursion
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
    
factorial(5)
```

```python
# filter palindrome from a list of string

def palindrome(word):
    rev=[]
    result=[]
    for i in word:
        rev.append(i[::-1])
    for j in word:
        if j in rev:
            result.append(j)

    return result

palindrome(['apple','sas','cow','151'])
```

```python
#Given a string, write a function to check if it is an anagram of another string.
def are_anegrams(s1,s2):
    return sorted(s1)==sorted(s2)

str1='silent'
str2='listen'

are_anegrams(str1,str2)
```
