## Drop column with 90% null values

```python
df=df.drop(columns=['feat1'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns= [col.lower().strip().replace(' ','_') for col in df.columns]
```

```python
df.columns
```

```python
str_cols= df.select_dtypes(include='object').columns.tolist()
```

```python
str_cols
```

```python
for col in str_cols:
    df[col]=df[col].str.lower().str.strip()
```

```python
df.head()
```

## checking for inconsitency in value in categorical columns

```python
for col in str_cols:
    print(df[col].value_counts())
```

```python
df['gender']=df['gender'].replace({'femle':'female'})
```

```python
df['gender'].value_counts()
```

```python
df.columns
```

```python
df['fbs']=df['fbs'].fillna(df['fbs'].median())
```

```python
df['hba1c']=df['hba1c'].fillna(df['hba1c'].median())
```

```python
df['blood_pressure']=df['blood_pressure'].fillna(df['blood_pressure'].mode()[0])
df['diet']=df['diet'].fillna(df['diet'].mode()[0])
```

```python
df.isna().sum()
```

```python
str_cols.pop()
```

```python
str_cols
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
for col in str_cols:
    df[col]=label_encoder.fit_transform(df[col])
```

```python
df.head()
```

```python
from sklearn.model_selection import cross_val_score,KFold,GridSearchCV
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns=['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
svc=SVC(kernel='sigmoid')
```

```python
kf=KFold(n_splits=5,shuffle=True,random_state=42)
dt_scores=cross_val_score(svc,X,y,cv=kf)
dt_mean_scores=np.mean(dt_scores)
```

```python
dt_mean_scores
```

```python
dt_scores
```

## Hyperparameter Tuning

```python
model_param={
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1.0,1.5,2.0],
    'tol':[0.003,0.002,0.001]
}
```

```python
model_grid_search=GridSearchCV(svc,model_param,cv=kf,scoring='accuracy')
model_grid_search.fit(X,y)
```

```python
model_grid_search.best_estimator_
```

```python
model_grid_search.best_params_
```

```python
model_grid_search.best_score_
```

```python
final_model= SVC(C=1.0,kernel='linear',tol=0.003)
```

```python
final_model.fit(X,y)
```
