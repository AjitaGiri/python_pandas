# Titanic Dataset Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('titanic.csv')
dataset.head()
```

```python
dataset.info()
```

## Descriptive Statistics

```python
dataset['Age']
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(subset=['Age'],inplace=True)
```

```python
mn=np.mean(dataset['Age'])
mn
```

```python
dataset['Age'].mean()
```

```python
sns.histplot(x='Age',data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],c='red')
plt.show()
```

```python
dataset['Age'].median()
```

```python
dataset['Fare'].median()
```

```python
dataset['Fare'].mean()
```

```python
dataset['Fare'].mode()[0]
```

```python
sns.histplot(x='Age',data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([dataset['Fare'].mean() for i in range(0,300)],[i for i in range(0,300)],c='red')
plt.plot([dataset['Fare'].median() for i in range(0,300)],[i for i in range(0,300)],c='blue')
plt.plot([dataset['Fare'].mode() for i in range(0,300)],[i for i in range(0,300)],c='green')

plt.show()
```
