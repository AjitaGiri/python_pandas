### Measure of Dispersion

```python
min_r=dataset['Age'].min()
```

```python
max_r=dataset['Age'].max()
```

```python
range_r =max_r-min_r
print(range_r)
```

Mean absolute deviation: it is the average distance between each data points and the mean.

```python
sec_a=np.array([75,65,73,68,72,67])
sec_b=np.array([90,47,43,96,93,51])
```

```python
mean_a=np.mean(sec_a)
mean_b=np.mean(sec_b)
```

```python
mad_a=np.sum(np.abs(sec_a-mean_a))/len(sec_a)
mad_a
```

```python
mad_b=np.sum(np.abs(sec_b-mean_b))/len(sec_b)
mad_b
```

Choose the data with low mean absolute deviation.
